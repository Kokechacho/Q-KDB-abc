```q
csvPath:"Data/IBEX.csv"
trades: ("DFFFFFJ"; enlist csv) 0: hsym `$csvPath;
```

```q
delete from trades where Open = 0N
days10: {[tab] select avgOpen:avg(Open), maxHigh: max(High), minLow: min(Low), avgClose: avg(Close), totalVolume: sum(Volume) by 10 xbar Date from tab}
show trades10: days10 trades
```

```q
exec avgOpen cor totalVolume from trades10
```
